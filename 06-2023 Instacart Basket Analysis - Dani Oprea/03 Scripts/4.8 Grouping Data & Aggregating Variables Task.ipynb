# 03. Finding The Aggregated Mean

```python
#Step 2 & 3

# Find the mean of the order_number column group by department_id

ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

Some of the means from the department_ids are pretty similar between the subset and the whole dataframe. For example, the mean of department_id 1 is 15.577 in the subset and 15.458 in the whole dataframe. However, some differ by a decent amount. For example, the mean of department_id 21 is 25.535 in the subset and 22.902 in the whole dataframe.

# 04. Creating A Loyalty Flag

```python
# Step 4

# Create the max_order column

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# Derive columns using the loc() function

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

# 05. Checking Spending Habits

```python
# Use the groupby function to display the mean, min and max prices for each loyalty flag

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

It looks like Loyal Customers, on average, spend less than either New customers and Regular customers.

# 06. Create a Spending Flag

```python
# Step 6

# Create the average_spending column

ords_prods_merge['average_spending'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(50)
```

```python
# Derive spending_flag columns using the loc() function

ords_prods_merge.loc[ords_prods_merge['average_spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spending'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check the frequency of the spending_flag column

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Return the first 60 rows of the user_id, average_spending, spending_flag, and prices columns

ords_prods_merge[['user_id', 'average_spending', 'spending_flag', 'prices']].head(60)
```

# 07. Create an Order Frequency Flag

```python
# Step 7

# Create the median_days column

ords_prods_merge['median_days'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
pd.options.display.max_rows = None
pd.options.display.max_columns = None

ords_prods_merge.head(100)
```

```python
# Derive order_frequency_flag columns using the loc() function

ords_prods_merge.loc[ords_prods_merge['median_days'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days'] <= 20) & (ords_prods_merge['median_days'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head(100)
```

```python
# Check the frequency of the order_frequency_flag column

ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Create a subset to view the missing values

df_nan = ords_prods_merge[ords_prods_merge['order_frequency_flag'].isnull() == True]
```
