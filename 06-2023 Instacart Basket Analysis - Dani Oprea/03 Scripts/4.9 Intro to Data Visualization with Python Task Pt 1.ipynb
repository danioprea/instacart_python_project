# 05. Data Consistency Checks

```python
# Check for mixed types

for col in cust.columns.tolist():
  weird = (cust[[col]].applymap(type) != cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (cust[weird]) > 0:
    print (col)
```

```python
# Convert to string type

cust['First Name'] = cust['First Name'].astype('str')
```

```python
# Recheck to see if the First Name column has been fixed

for col in cust.columns.tolist():
  weird = (cust[[col]].applymap(type) != cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (cust[weird]) > 0:
    print (col)
  else:
    print(col + ' is all clear')
```

We have fixed any mixed type issues.

```python
# Check for missing values

cust.isnull().sum()
```

There are no missing values.

```python
# Finding duplicates

df_dups = cust[cust.duplicated()]
```

```python
# Check to see if there are any duplicate values

df_dups
```

# 06. Merging Customer Data with Main Instacart Data

```python
# Import orders_products_merged.pkl

ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.shape
```

```python
# We experienced issues later on in the notebook when trying to merge the dataframes so we're going back to check the
# types

ords_prods_merge.dtypes
```

It looks like user_id in the ords_prods_merge dataframe is int64 instead of object type, so let's update that.

```python
# Update variable type of user_id column in ords_prods_merge to object so that we can merge with cust dataframe

ords_prods_merge['user_id'] = ords_prods_merge['user_id'].astype('str')
```

```python
# Recheck data types

ords_prods_merge['user_id'].dtype
```

```python
# Remove _merge column from ords_prods_merge

ords_prods_merge_1 = ords_prods_merge.drop('_merge', axis=1)
```

```python
# Check

ords_prods_merge_1.head()
```

```python
# Merge the dataframes into a new dataframe df_merged using user_id as a key

df_merged_1 = cust.merge(ords_prods_merge_1, on = 'user_id', indicator = True)
```
